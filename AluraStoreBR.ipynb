## 4. ANÁLISES OBRIGATÓRIAS

### 4.1 — Faturamento Total das Lojas

```python
# ============================================
# 4.1. FATURAMENTO TOTAL DAS LOJAS
# ============================================

faturamentos = {}  # dicionário para armazenar resultado

for nome, df in lojas.items():
    # Somar o faturamento total da loja
    faturamentos[nome] = df["Preço"].sum()

# Exibir o resultado final
faturamentos
```

### 4.2 — Categorias mais e menos vendidas

#### a) Contar vendas por categoria

```python
# ============================================
# 4.2. CATEGORIAS MAIS E MENOS VENDIDAS
# ============================================

vendas_categorias = {}

for nome, df in lojas.items():
    # Contagem de vendas por categoria
    categorias = df['Categoria do Produto'].value_counts()
    vendas_categorias[nome] = categorias

vendas_categorias
```

#### b) Identificar a categoria mais vendida e a menos vendida por loja

```python
# ============================================
# 4.2. CATEGORIAS MAIS E MENOS VENDIDAS (COM QUANTIDADE)
# ============================================

vendas_categorias = {}

for nome, df in lojas.items():
    # Contagem de vendas por categoria
    categorias = df['Categoria do Produto'].value_counts()
    vendas_categorias[nome] = categorias


categorias_extremos = {}

for nome, categorias in vendas_categorias.items():

    cat_mais = categorias.idxmax()
    qtd_mais = categorias.max()

    cat_menos = categorias.idxmin()
    qtd_menos = categorias.min()

    categorias_extremos[nome] = {
        "Categoria Mais Vendida": cat_mais,
        "Qtd Mais Vendida": qtd_mais,
        "Categoria Menos Vendida": cat_menos,
        "Qtd Menos Vendida": qtd_menos
    }
```

#### c) Unificar tudo em uma tabela final (DataFrame)

```python
categorias_df = pd.DataFrame(categorias_extremos).T
categorias_df
```

### 4.3. Média das avaliações por loja

#### a) Calcular a média de avaliação por loja

```python
# ============================================
# 4.3. MÉDIA DAS AVALIAÇÕES POR LOJA
# ============================================

media_avaliacoes = {}

for nome, df in lojas.items():
    media = df['Avaliação da compra'].mean()
    media_avaliacoes[nome] = round(media, 2)
```

#### b) Transformar em DataFrame

```python
avaliacoes_df = pd.DataFrame(
    list(media_avaliacoes.items()),
    columns=['Loja', 'Média de Avaliação']
)

avaliacoes_df
```

### 4.4. Produtos mais e menos vendidos

#### a) Agrupar por produto e contar quantidade vendida

```python
# ============================================
# 4.4. PRODUTOS MAIS E MENOS VENDIDOS
# ============================================

vendas_produtos = {}

for nome, df in lojas.items():
    produtos = df['Produto'].value_counts()
    vendas_produtos[nome] = produtos
```

#### b) Identificar o produto mais vendido e o menos vendido

```python
produtos_extremos = {}

for nome, produtos in vendas_produtos.items():

    prod_mais = produtos.idxmax()
    qtd_mais = produtos.max()

    prod_menos = produtos.idxmin()
    qtd_menos = produtos.min()

    produtos_extremos[nome] = {
        "Produto Mais Vendido": prod_mais,
        "Qtd Vendida (Mais)": qtd_mais,
        "Produto Menos Vendido": prod_menos,
        "Qtd Vendida (Menos)" : qtd_menos
    }
```

#### c) Unificar resultados em um DataFrame final

```python
produtos_df = pd.DataFrame(produtos_extremos).T
produtos_df
```

### 4.5. Frete Médio por Loja

#### a) Calcular o frete médio por loja

```python
# ============================================
# 4.5. FRETE MÉDIO POR LOJA
# ============================================

frete_medio = {}

for nome, df in lojas.items():
    media_frete = df['Frete'].mean()
    frete_medio[nome] = round(media_frete, 2)
```

#### b) Transformar em DataFrame

```python
frete_df = pd.DataFrame(
    list(frete_medio.items()),
    columns=['Loja', 'Frete Médio (R$)']
)

frete_df
```

## 5. CRIAÇÃO DAS VISUALIZAÇÕES

### 5.1. Gráfico — Faturamento Total por Loja

```python
# ============================================
# Cálculo do faturamento total por loja
# ============================================

faturamento = {}

for nome, df in lojas.items():
    # Considerando faturamento = preço * 1 venda (uma linha por venda)
    total = df['Preço'].sum()
    faturamento[nome] = round(total, 2)

faturamento_df = pd.DataFrame(
    list(faturamento.items()),
    columns=['Loja', 'Faturamento Total (R$)']
)

faturamento_df
```

```python
plt.bar(faturamento_df['Loja'], faturamento_df['Faturamento Total (R$)'])

# adicionar valores
for i, v in enumerate(faturamento_df['Faturamento Total (R$)']):
    plt.text(i, v + (v * 0.01), f"R${v:,.2f}", ha='center', fontsize=10)

plt.title("Faturamento Total por Loja")
plt.xlabel("Loja")
plt.ylabel("Faturamento (R$)")
plt.xticks(rotation=0)
plt.show()
```

### 5.2. Gráfico: — Categorias mais vendidas (por loja)

#### a) Distribuição das Categorias por loja

```python
# ============================================
# Gráficos de pizza – Categorias por loja
# ============================================

for nome, categorias in vendas_categorias.items():
    plt.figure(figsize=(6,6))
    plt.pie(
        categorias,
        labels=categorias.index,
        autopct='%1.1f%%',
        startangle=90
    )
    plt.title(f"Distribuição das Categorias – {nome}")
    plt.show()
```

#### b) Consolidado das categorias (Todas as lojas)

```python
# ============================================
# Pizza consolidada das categorias (todas as lojas)
# ============================================

# concatenando todas as lojas em um único DataFrame
todas_as_lojas = pd.concat(lojas.values())

# contando vendas por categoria no geral
categorias_consolidadas = todas_as_lojas['Categoria do Produto'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(
    categorias_consolidadas,
    labels=categorias_consolidadas.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Categorias de Produtos – Consolidado das 4 Lojas")
plt.show()
```

### 5.3. Gráfico — Avaliação Média

```python
plt.bar(avaliacoes_df['Loja'], avaliacoes_df['Média de Avaliação'])

# adicionar valores
for i, v in enumerate(avaliacoes_df['Média de Avaliação']):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontsize=10)

plt.title("Média de Avaliação por Loja")
plt.xlabel("Loja")
plt.ylabel("Nota Média")
plt.show()
```

### 5.4. Gráfico — Frete Médio

```python
plt.bar(frete_df['Loja'], frete_df['Frete Médio (R$)'])
# adicionar valores
for i, v in enumerate(frete_df['Frete Médio (R$)']):
    plt.text(i, v + 0.1, f"R${v:.2f}", ha='center', fontsize=10)

plt.title("Frete Médio por Loja")
plt.xlabel("Loja")
plt.ylabel("Média do Frete (R$)")
plt.show()
```

### 5.5. Gráfico — Top 5 Produtos mais vendidos

```python
# ============================================
# Top 5 produtos mais vendidos – gráfico horizontal
# ============================================

for nome, produtos in vendas_produtos.items():
    top5 = produtos.head(5)

    plt.barh(top5.index, top5.values)
    # adicionar valores ao lado das barras
    for i, v in enumerate(top5.values):
        plt.text(v + 0.5, i, str(v), va='center', fontsize=10)

    plt.title(f"Top 5 Produtos Mais Vendidos – {nome}")
    plt.xlabel("Quantidade Vendida")
    plt.ylabel("Produto")
    plt.gca().invert_yaxis()  # deixa o mais vendido no topo
    plt.show()
```

### 8.3. Heatmap simples de densidade (Matplotlib)

```python
# ============================================
# 8.3. HEATMAP SIMPLES DE DENSIDADE
# ============================================

import numpy as np

todas_lojas = pd.concat(lojas.values())

plt.figure(figsize=(8,6))

plt.hist2d(
    todas_lojas['lon'],
    todas_lojas['lat'],
    bins=50,
    cmap='hot'
)

plt.colorbar(label='Densidade de Vendas')
plt.title("Heatmap de Densidade – Vendas Consolidadas")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
